```python
### uploading the dataset

data=pd.read_csv('application_data.csv')
data.head(10)
```

```python
## understanding the dataset

data.shape
```

```python
## checking the null values present in the dataset
data.isnull().sum()
```

```python
## checking the percentage of the missing values present in the columns


data.isnull().sum()/len(data)*100

```

```python
## checking the missing values more than 30% in the columns

missing_column= data.isnull().sum()/len(data)*100
missing_column=missing_column[missing_column.values>0.3]
missing_column
len(missing_column)

```

```python
## dropping columns which have more than 30% missing values

missing_column=list(missing_column[missing_column.values>0.30].index)
data.drop(labels=missing_column,axis=1,inplace=True)
print(len(missing_column))
```

```python
data.isnull().sum()/len(data)*100
```

```python
## AMT_ANNUITY have very few missing values and it is one of the important column

data.AMT_ANNUITY.value_counts()
```

```python
### checking whether there is any outlier in 'AMT_ANNUITY' column:


data.boxplot(column='AMT_ANNUITY')
plt.show()
```

```python
## Outliers present in the above column is very large in numbers hence those values are to be replaced by median value.


median_value=data['AMT_ANNUITY'].median()

data.loc[data['AMT_ANNUITY'].isnull(),'AMT_ANNUITY']=median_value

data.isnull().sum()
```

```python
## 'AMT_GOODS_PRICE' column has very less numbers of null values hence those can be replaced by mean value.


mean_value=data['AMT_GOODS_PRICE'].mean()
mean_value

data.loc[data['AMT_GOODS_PRICE'].isnull(),'AMT_GOODS_PRICE']=mean_value

data.isnull().sum()
```

```python
# rows having more than 30% null values can be dropped

missing_rows=data.isnull().sum(axis=1)

missing_rows=list(missing_rows[missing_rows.values>=0.3*len(data)].index)

data.drop(labels=missing_rows,axis=0,inplace=True)

print(len(missing_rows))
```

```python
## Will drop the columns which are not important parametres for the analysis:


dropped_columns=['FLAG_MOBIL', 'FLAG_EMP_PHONE', 'FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE',
       'FLAG_PHONE', 'FLAG_EMAIL','REGION_RATING_CLIENT','REGION_RATING_CLIENT_W_CITY','FLAG_EMAIL','CNT_FAM_MEMBERS', 'REGION_RATING_CLIENT',
       'REGION_RATING_CLIENT_W_CITY','DAYS_LAST_PHONE_CHANGE', 'FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_3','FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5', 'FLAG_DOCUMENT_6',
       'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8', 'FLAG_DOCUMENT_9','FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11', 'FLAG_DOCUMENT_12',
       'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14', 'FLAG_DOCUMENT_15','FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17', 'FLAG_DOCUMENT_18',
       'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21']


data.drop(labels=dropped_columns,axis=1,inplace=True)
```

```python
##There are certain columns with 'XNA' values

### checking one of the important categorical column 'CODE_GENDER'

data.CODE_GENDER.value_counts()
```

```python
## There are 4 rows with 'XNA' values which means data not available.Since only 4 rows are missing out of 307511 rows and majority are females so we can replace 'XNA' with 'F'

data.loc[data['CODE_GENDER']=='XNA','CODE_GENDER']='F'
data['CODE_GENDER'].value_counts()
```

```python
## Creating bins for 'AMT_INCOME_TOTAL' column

bins=[0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,325000,350000,375000,400000,425000,450000,475000,500000,10000000000]
labels = ['0-25000', '25000-50000','50000-75000','75000,100000','100000-125000', '125000-150000', '150000-175000','175000-200000',
       '200000-225000','225000-250000','250000-275000','275000-300000','300000-325000','325000-350000','350000-375000',
       '375000-400000','400000-425000','425000-450000','450000-475000','475000-500000','500000 and above']

data['income_range']=pd.cut(data['AMT_INCOME_TOTAL'],bins=bins,labels=labels)

```

```python
## UNIVARIATE ANALYSIS FOR TARGET_0(Clients who have no payment difficulties)



def UnivariatePlot(data, col, title, xtitle, ytitle, hue = None):
    
    sns.set_style("white")
    sns.set_context("notebook")    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(data[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation = 45)
    plt.yscale('log')
    plt.title(title)

    ax = sns.countplot(data = data, x= col, order=data[col].value_counts().index,hue = hue, palette='bright') 
    ax.set(xlabel = xtitle, ylabel = ytitle)    
    plt.show()
    
    
UnivariatePlot(data_target_0, col='income_range', title='Distribution of Income', hue='CODE_GENDER', xtitle = "Income Range", ytitle = "Count")
```

Findings from the graph:

1.Income range between 120000-180000 have the higher credits and number of female members are more than male members in that range.

1.The above graph shows that the income type 'working', 'Commercial associate' and 'State Servant' have the highest credit and number of Female members are more than male members in that range.
